# 자료형 변경 

- 묵시적 형변환 : 자동 => 정수 < 실수 < 문자
- 강제적 형변환 
```
실수 => 정수 
실수, 정수 => 문자열 
문자열숫자 => 숫자
    pd.to_numeric()
날짜스타일 문자열 2021-07-31, 20211126  => 시계열 데이타 
    pd.to_datetime()
```

````python
# 10행 5열 
data = np.arange(1,51).reshape(-1 , 5)
columnsList = ['col'+str(i) for i in range(1,6)]
rowList = ['row'+str(i) for i in range(1,11)]
df1 = pd.DataFrame(data, columns=columnsList, index=rowList)
df1
````

````python
df1.dtypes
````

````python
# 실수형 데이타 삽입 => 값이 교체된 컬럼의 데이타형이 실수형으로 자동 변경 
df1.loc['row1', 'col1'] = 3.14
df1
````

````python
df1.dtypes
````

````python
# 문자형 데이타 삽입 => 값이 교체된 컬럼의 데이타형이 문자열형으로 자동 변경 
df1.loc['row1', 'col2'] = 'Python'
df1.loc['row1', 'col3'] = 'Python'
df1
````

````python
# 연속된 숫자문자 
data = np.arange(20210101, 20210112).astype(np.str)
data
````

````python
# 데이타프레임 생성 
df2 = pd.DataFrame(data, columns=['d1'])
df2
````

````python
df2.dtypes
````

````python
# pd.to_datetime(넘파이배열/시리즈, errors='coerce')
# d1 컬럼을 시계열데이타로 변경 => df2['d2']
pd.to_datetime(df2['d1'], errors='coerce')
````

````python
df2['d2'] = pd.to_datetime(df2['d1'], errors='coerce')
df2
````

````python
# 시계열데이타는 자료형 datetime64[ns]
df2.dtypes
````

````python
# 시계열 데이타 분리 
# dt.옵션 지정해서 년,월,일,요일 로 분리 가능 
# dt.month, dt.year, dt.day, dt.weekday (0~6 0은 월요일 )
````

````python
# 요일 분리 
df2['d2'].dt.weekday
````

````python
# d2 컬럼에서 년,월,일,요일 분리해서 새로운 컬럼으로 생성 
df2['d3'] = df2['d2'].dt.year
df2['d4'] = df2['d2'].dt.month
df2['d5'] = df2['d2'].dt.day
df2['d6'] = df2['d2'].dt.weekday
````

# 컬럼 값을 특정 조건에 따라 교체하기 
```
예시) 

0 => mon , 1 => tue, ... 
남자 => 1, 여자 => 2 ... 
강서구 방화동 => 방화동 
```
- for + if 등의 제어문 
- map({old1:new1, old2:new2 ... }) 
- replace({old1:new1, old2:new2 ... })
- replace([old1, old2 ...], [new1,new2 ...] })
- apply(람다함수/함수)


````python
# df2['d6'] 컬럼 복제 
df2['d7'] = df2['d6']
df2
````

````python
# d7 컬럼에서 mon, tue, wed ... 
# replace({old1:new1, old2:new2 ... })
df2['d7'].replace({ 0:'mon', 1:'tue', 2:'wed', 3:'thu', 4:'fri', 5:'sat', 6:'sun'})
````

````python
# 원본 컬럼에 적용 
df2['d7'] = df2['d7'].replace({ 0:'mon', 1:'tue', 2:'wed', 3:'thu', 4:'fri', 5:'sat', 6:'sun'})
df2
````

````python
# apply() 함수를 이용해서 값 교체하기 

# apply() 에 적용되는 함수 정의 
def change_day(x):
    week_day = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']
    if 0 <= x <= 6 :
        return week_day[x]
    else:
        return '잘못된 입력'
````

````python
change_day(0), change_day(5)
````

````python
change_day(10)
````

````python
# apply 함수 적용 
df2['d6'].apply(change_day)
````

````python
# d6 컬럼의 apply() 적용해서 새로운 컬럼 생성 
df2['d8'] = df2['d6'].apply(change_day)
````
