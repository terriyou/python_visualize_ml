````python
# Python 샘플 코드 #
import requests

# 1) 데이타 요청 
myServiceKey = 'Yhq6n9vlpaKZsCOw0jec8zIdZ8p+Bpuku6WLjVgDhqXRW6dHnfXoauSEj19jpnjv59CUGspyTxVCgCeXOua7dg=='
url = 'http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson'

params ={'serviceKey' : myServiceKey, 
         'pageNo' : '1', 'numOfRows' : '10', 
         'startCreateDt' : '20210101', 
         'endCreateDt' : '20211129' }

response = requests.get(url, params=params)
print(response.content)
````

````python
# xml 데이타 => 텍스트 xml 데이타 
xml_string = response.text
# response.content
type(xml_string), type(response.content)
````

````python
# 텍스트 xml 데이타  => OrderedDict 
orderedDict_data = xmltodict.parse(xml_string)
print(type(orderedDict_data))

# OrderedDict => json 텍스트 => 딕셔너리 
dict = json.loads(json.dumps(orderedDict_data))
print(type(dict))
````

````python
# 딕셔너리 => 리스트 
covid_list = dict['response']['body']['items']['item']
covid_list
````

````python
# 리스트딕셔너리 => 데이타프레임 
df_covid = pd.DataFrame(covid_list)
df_covid
````

````python
# 기준일	STATE_DT
# 확진자 수	DECIDE_CNT
# 격리해제 수	CLEAR_CNT
# 검사진행 수	EXAM_CNT
# 사망자 수	DEATH_CNT

df_covid2 = df_covid[['stateDt', 'decideCnt', 'clearCnt', 'examCnt', 'deathCnt']]
df_covid2
````
