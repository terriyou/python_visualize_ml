```
4. 2000년도 이후데이터 에 대해  출생량 을 구하고 
남아수가 1800000 이상이면 '많음'아니면 '적음' 이라고 출력하여라


	남아수 출생량
2000   1814601   '많음'
2001   1799049   '적음'
```

````python
after_2000 = birthArr[birthArr[:,0] > 2000]
print(after_2000)
````

````python
result = np.where(after_2000[:,1]>=1800000,'많음','적음')
````

````python
result = result.reshape(11,1)
````

````python
str_aft_2000 = after_2000.astype(dtype='<U10')
````

````python
str_aft_2000_added = np.append(str_aft_2000, result, axis=1)
````

````python
print(str_aft_2000_added)
````

````python
b_a_1800 = birthArr[(birthArr[:,0] >= 1800) & (birthArr[:,0] < 1900)]
b_a_1900 = birthArr[(birthArr[:,0] >= 1900) & (birthArr[:,0] < 2000)]
b_a_2000 = birthArr[(birthArr[:,0] >= 2000) & (birthArr[:,0] < 2100)]
````

````python
boy_1800 = b_a_1800[:,1]
boy_mean_1800 = boy_1800.mean()
girl_1800 = b_a_1800[:,2]
girl_mean_1800 = girl_1800.mean()
````

````python
print(boy_mean_1800)
print(girl_mean_1800)
````

````python
boy_1900 = b_a_1900[:,1]
boy_mean_1900 = boy_1900.mean()
girl_1900 = b_a_1900[:,2]
girl_mean_1900 = girl_1900.mean()
````
