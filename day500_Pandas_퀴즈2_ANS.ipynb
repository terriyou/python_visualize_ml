## ANS : Q1-5

````python
# 확진일 컬럼값을 아래와 같은 형태로 변경하여라. 
# '1.24.' => '2020-1-24'
````

````python
def change(x):
    # 2020- + .를 - 로 대체     
    result = '2020-'+x.replace('.','-')
    # 마지막 . 제외하고 나머지를 리턴 
    return result[:-1]
````

````python
change('1.24.')
````

````python
df_covid['확진일'] = df_covid['확진일'].apply(change)
````

````python
df_covid
````
