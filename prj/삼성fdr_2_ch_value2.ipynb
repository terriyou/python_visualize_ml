```python
start = ['2020-11-02','2020-11-09','2020-11-16','2020-11-23','2020-11-30','2020-12-07']
final = ['2021-11-03','2021-11-10','2021-11-17','2021-11-24','2021-12-01','2021-12-08']
param = [(1,1,0),(1,2,0),(2,1,0),(2,2,0),(3,1,0),(3,2,0),(4,1,0),(4,2,0)]
```

```python
for i in range(len(start)):
  samsung_df = fdr.DataReader('005930', f'{start[i]}', f'{final[i]}')
  samsung_df = samsung_df[["Close"]].astype(float)
  samsung_df = samsung_df.reset_index()
  samsung_df.columns = ['day', 'price']
  samsung_df['day'] = pd.to_datetime(samsung_df['day'])
  samsung_df.index = samsung_df['day']
  samsung_df.set_index('day', inplace=True)
  samsung_train_df = samsung_df[:-5]
  samsung_test_df = samsung_df[-5:]
  for j in range(len(param)):
    model = ARIMA(samsung_train_df.price.values, order = param[j])
    model_fit = model.fit(trend = 'c', full_output = True, disp = True)

    forecast_data = model_fit.forecast(steps=5)

    pred_arima_y = forecast_data[0].tolist()

    test_y = samsung_test_df.price.values
    pred_y_lower = []
    pred_y_upper = []

    for lower_upper in forecast_data[2]:
      lower = lower_upper[0]
      upper = lower_upper[1]
      pred_y_lower.append(lower)
      pred_y_upper.append(upper)
    plt.figure(figsize=(10, 7))

    plt.plot(pred_arima_y, color = 'gold')
    plt.plot(pred_y_lower, color = 'red')
    plt.plot(pred_y_upper, color = 'blue')
    plt.plot(test_y, color = 'green')

    plt.legend(['prediction','lower','upper','real'])
    plt.title(f"{param[j]}")
    plt.show()
```

```python
samsung_df = fdr.DataReader('005930', '2020-12-14', '2021-12-15')
samsung_df = samsung_df[["Close"]].astype(float)
samsung_df = samsung_df.reset_index()
samsung_df.columns = ['day', 'price']
samsung_df['day'] = pd.to_datetime(samsung_df['day'])
samsung_df.index = samsung_df['day']
samsung_df.set_index('day', inplace=True)
samsung_train_df = samsung_df
samsung_train_df
plot_acf(samsung_df.price.values)
plot_pacf(samsung_df.price.values)
plt.show()
```

```python
model = ARIMA(samsung_train_df.price.values, order = (1,2,0))
model_fit = model.fit()
print((1,2,0))
print(model_fit.summary())
```

```python
forecast_data = model_fit.forecast(steps=5)

pred_arima_y = forecast_data[0].tolist()


pred_y_lower = []
pred_y_upper = []

for lower_upper in forecast_data[2]:
  lower = lower_upper[0]
  upper = lower_upper[1]
  pred_y_lower.append(lower)
  pred_y_upper.append(upper)
plt.figure(figsize=(10, 7))

plt.plot(['2021-12-15','2021-12-16','2021-12-17','2021-12-18','2021-12-19','2021-12-20'],[samsung_train_df.iloc[-1,0].tolist()] + pred_arima_y, color = 'gold')
plt.plot(['2021-12-15','2021-12-16','2021-12-17','2021-12-18','2021-12-19','2021-12-20'],[samsung_train_df.iloc[-1,0].tolist()] + pred_y_lower, color = 'red')
plt.plot(['2021-12-15','2021-12-16','2021-12-17','2021-12-18','2021-12-19','2021-12-20'],[samsung_train_df.iloc[-1,0].tolist()] + pred_y_upper, color = 'blue')

plt.annotate(str(int(samsung_train_df.iloc[-1,0].tolist())), \
             xy=(0,int(samsung_train_df.iloc[-1,0].tolist())), xytext=(0,75000), fontsize=14, ha='center',\
             arrowprops=dict(facecolor='black', width=1, shrink=0.1, headwidth=10))

plt.annotate(str(int(pred_arima_y[0])), xy=(1,pred_arima_y[0]), xytext=(1,75000), fontsize=14, ha='center',\
             arrowprops=dict(facecolor='black', width=1, shrink=0.1, headwidth=10))
plt.annotate(str(int(pred_arima_y[1])), xy=(2,pred_arima_y[1]), xytext=(2,75000), fontsize=14, ha='center',\
             arrowprops=dict(facecolor='black', width=1, shrink=0.1, headwidth=10))
plt.annotate(str(int(pred_arima_y[2])), xy=(3,pred_arima_y[2]), xytext=(3,75000), fontsize=14, ha='center',\
             arrowprops=dict(facecolor='black', width=1, shrink=0.1, headwidth=10))
plt.annotate(str(int(pred_arima_y[3])), xy=(4,pred_arima_y[3]), xytext=(4,75000), fontsize=14, ha='center',\
             arrowprops=dict(facecolor='black', width=1, shrink=0.1, headwidth=10))
plt.annotate(str(int(pred_arima_y[4])), xy=(5,pred_arima_y[4]), xytext=(5,75000), fontsize=14, ha='center',\
             arrowprops=dict(facecolor='black', width=1, shrink=0.1, headwidth=10))

plt.legend(['prediction','lower','upper','real'])
plt.title('Samsung Stock Close Price Prediction')
plt.show()
```
